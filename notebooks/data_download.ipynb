```python
from core.data_engine import DataEngine  # 修正导入错误，应该是DataEngine而不是DataLoader


def main():
    # 定义测试参数
    symbols = ["AAPL", "GOOGL", "MSFT", "TSLA"]
    start_date = "2023-01-01"
    end_date = "2023-12-31"

    # 创建数据引擎实例
    engine = DataEngine(symbols)

    print("开始下载股票数据...")

    # 使用数据引擎批量下载数据
    engine.update_all_data(start_date, end_date)

    print(f"\n成功下载了 {len(engine.universe_data)} 只股票的数据")

    # 显示已有的数据文件
    available_files = engine.loader.get_available_data()
    print("\n本地已有的数据文件: ")
    for file in available_files:
        print(f"- {file}")


if __name__ == "__main__":
    main()
```
